# 02 이것저것

```python
pretrained_model = "cosmoquester/bart-ko-small"
pretrained_tokenizer = "cosmoquester/bart-ko-small"


tokenizer = AutoTokenizer.from_pretrained(pretrained_tokenizer)
model = TFAutoModelForSeq2SeqLM.from_pretrained(pretrained_model, use_cache=False)
```

```python
model.summary()
```

```python
tokenizer.tokenize("육군 복무신조 우리의 결의 우리는 국가와 국민에 충성을 다하는 대한민국 육군이다")
```

```python
batch_size = 1
```

```python
def get_chat_data():
  for question, answer in zip(lyric_list, title_list):
    bos_token = [tokenizer.bos_token_id]
    eos_token = [tokenizer.eos_token_id]
    sent = tokenizer.encode('<lyric>' + question + '<title>' + answer) 
    yield bos_token + sent + eos_token
```

```python
dataset = tf.data.Dataset.from_generator(get_chat_data, output_types=tf.int32)
dataset = dataset.padded_batch(batch_size=batch_size, padded_shapes=(None,), padding_values=tokenizer.pad_token_id)
```

```python
for batch in dataset:
    print(batch)
    break
```

```python
adam = tf.keras.optimizers.Adam(learning_rate=3e-5, epsilon=1e-08)
steps = len(df) // batch_size
print("steps : ",steps, ", len(df) : ",len(df))
```

## 가 gradient tape 설정

```python
EPOCHS = 10


for epoch in range(EPOCHS):
  epoch_loss = 0

  for batch in tqdm.notebook.tqdm(dataset, total=steps):
      with tf.GradientTape() as tape:
          result = model(batch, labels=batch)  # result랑 loss 값 확인
          loss = result[0]
          batch_loss = tf.reduce_mean(loss)
          
      grads = tape.gradient(batch_loss, model.trainable_variables)
      adam.apply_gradients(zip(grads, model.trainable_variables))
      epoch_loss += batch_loss / steps

  print('[Epoch: {:>4}] cost = {:>.9}'.format(epoch + 1, epoch_loss))
  tf.saved_model.save(model, '/content/drive/MyDrive/trans_koreng_project/kogpt2/guk_KoBART_{0}_hungry.h5'.format(epoch + 1))
```
